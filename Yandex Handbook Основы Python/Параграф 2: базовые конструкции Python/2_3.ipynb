Формат ввода
Вводится одно натуральное число большее единицы.

Формат вывода
Требуется составить математическое выражение – произведение простых неубывающих чисел, которое в результате дает исходное.

```python
N = int(input())
count = 2
while N != 1:
    if N % count == 0:
        if N == count:
            print(count)
        else:
            print(f"{count} * ", end="")
        N = N // count
    else:
        count += 1
```

Блокчейн действительно представляет собой последовательность блоков.
```python
N = int(input())
h_n1 = 0
result = -1
for i in range(N):
    bn = int(input())
    m = bn // (256 * 256)
    r = (bn - m * 256 * 256) // 256
    h = bn - m * 256 * 256 - r * 256 
    if h >= 100 or h != 37 * (m + r + h_n1) % 256:
        result = i 
        break
    h_n1 = h
print (result)
```
