2.2. Рейтинг ученика
Продолжаем разбираться с финальной задачей блока. Первый шаг позади, вы молодец!

Вы уже умеете считывать критичные параметры учебной программы. Теперь пришло время собрать программу, которая показывала бы, насколько успешно учатся ученики.

Что нужно сделать
Ваша задача — проанализировать вовлечённость ученика и вычислить его рейтинг. А если входные данные окажутся неверными — указать это.

Формат ввода
Одна строчка, в которой через пробел подаются следующие данные:
N M Q cw sw hw tw
N (num_students) — число учеников на курсе
M (num_lessons) — количество занятий в курсе
Q (max_rating) — максимальный рейтинг
cw (classwork_coefficient) — коэффициент для активности на занятии
sw (selfwork_coefficient) — коэффициент для практики
hw (homework_coefficient) — коэффициент для домашней работы
tw (testwork_coefficient) — коэффициент для контрольной работы

Далее идёт фамилия ученика и несколько строчек (M) с оценками ученика на занятиях в формате:
Иванов
a₁,b₁,c₁,d₁  
a₂,b₂,c₂,d₂  
a₃,b₃,c₃,d₃  
⋮  
aₘ,bₘ,cₘ,dₘ
Где aᵢ,bᵢ,cᵢ,dᵢ — это оценки за i-ое занятие:

aᵢ (classwork_grade) — оценка за активность на занятии
bᵢ (selfwork_grade) — оценка за практику на занятии
cᵢ (homework_grade) — оценка за домашнюю работу на занятии
dᵢ (testwork_grade) — оценка за контрольную работу на занятии

Если не было активности, ученик отсутствовал либо не проявил себя на занятии — значение за активность будет 0.

Формат вывода
Фамилия и процент от максимального рейтинга, округленный до целого числа с помощью функции round(). Через пробел.
Иванов x%


```python
N, M, Q, cw, sw, hw, tw = map(int, input().split())
if M <= 0 or cw <= 0 or sw <= 0 or hw <= 0 or tw <= 0:
    name = input()
    print("Во введённых данных ошибка")
else:
    rating = 0
    for i in range(N):
        name = input()
        for i in range(M):
            a, b, c, d = map(int, input().split(","))
            rating += a * cw + b * sw + c * hw + d * tw
    if rating > Q:
        print("Во введённых данных ошибка")
    else:
        print(f"{name} {round(rating / Q * 100)}%")

```

Округлить результаты до целого значения с помощью функции round().
Вывести лучших учеников: топ-3 лидера по рейтингу, их фамилии в порядке убывания рейтинга и заработанный процент от максимального количества баллов. Если у студентов равный рейтинг (в абсолютном значении), то они идут в порядке их ввода — кто первый указан, тот выше по рейтингу и находится. Фамилии учеников уникальны.
На основе данных сделать предварительный вывод, насколько усваивается курс:
Если среднее значение рейтинга <= 50 — курс усваивается плохо.
Если среднее значение рейтинга > 50 — курс усваивается хорошо.

```python
N, M, Q, cw, sw, hw, tw = map(int, input().split())
top1_name = ""
top1_rating = -1

top2_name = ""
top2_rating = -1

top3_name = ""
top3_rating = -1

flag = True

if M <= 0 or cw <= 0 or sw <= 0 or hw <= 0 or tw <= 0 or N < 3: #проверки из условия
    flag = False
else:
    rating = 0
    max_rating = 0
    min_rating = 1000000000
    avg_rating = 0
    for i in range(N): #цикл по ученикам
        name = input()
        rating = 0
        for j in range(M): #цикл по учебным дням
            a, b, c, d = map(int, input().split(","))
            rating += a * cw + b * sw + c * hw + d * tw #формула вычсиления рейтинга
            
        if rating > top1_rating: #проверка на топ 1 рейтинг
            top3_rating = top2_rating
            top3_name = top2_name

            top2_rating = top1_rating
            top2_name = top1_name

            top1_rating = rating
            top1_name = name

        elif rating > top2_rating: #проверка на топ 2
            top3_rating = top2_rating
            top3_name = top2_name

            top2_rating = rating
            top2_name = name

        elif rating > top3_rating: #проверка на топ 3
            top3_rating = rating
            top3_name = name
                
        if rating > Q:
            flag = False
        else:
            if rating > max_rating:
                max_rating = rating
            if rating < min_rating:
                min_rating = rating
            avg_rating += rating #расчет суммы рейтинга
    if flag:
        print(round(max_rating / Q * 100), round(avg_rating / (N * Q) * 100), round(min_rating / Q * 100))
        print(f"{top1_name} {round(top1_rating / Q * 100)}%")
        print(f"{top2_name} {round(top2_rating / Q * 100)}%")
        print(f"{top3_name} {round(top3_rating / Q * 100)}%")
        if round(avg_rating / (Q * N) * 100) > 50:
            print("Курс усваивается хорошо")
        else:
            print("Курс усваивается плохо")
    else:
        print("Во введённых данных ошибка")
```
