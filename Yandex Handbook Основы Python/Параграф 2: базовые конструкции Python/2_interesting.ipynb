1) Общая сумма цифр введенных чисел

```python
N = int(input())
sum_a = 0
for i in range(N):
    a = int(input())
    while a:
        sum_a += a % 10
        a = a // 10
print(sum_a)
```

2) Нахождения НОДа N чисел

```python
N = int(input())
b = int(input())
for i in range(N - 1):
    a = int(input())
    a1 = max(a, b)
    b1 = min(a, b)
    r = a1 % b1
    if (r == 0):
        b = b1
    else:
        while r != 0:
            a1 = b1
            b1 = r 
            r = a1 % b1
        b = b1
print(b)
```

3) Максимальная цифра в каждом введенном числе

```python
N = int(input())

for i in range(N):
    num_n = int(input())
    max_digit = 0
    while num_n:
        digit = num_n % 10
        if digit > max_digit:
            max_digit = digit
        num_n //= 10
    print(max_digit)
```

4) Палиндром - считаем количество введенных чисел - палиндромов

```python
N = int(input())
count = 0
for i in range(N):
    k = 0
    numb = int(input())
    numb_base = numb
    while numb:
        k = 10 * k + numb % 10
        numb //= 10
    if k == numb_base:
        count += 1   
print(count)
```

5) Выводим систему исчесления, в которой число имеет наибольшую сумму цифр

```python
N = int(input())
max_i = 0
max_total = 0
for i in range(2, 11):
    total = 0
    N1 = N
    while N1:
        total += N1 % i
        N1 //= i
    if total > max_total:
        max_i = i
        max_total = total
print(max_i)
```

6) Основная теорема арифметики - выводим разложение числа на простые множители 

```python
N = int(input())
count = 2
while N != 1:
    if N % count == 0:
        if N == count:
            print(count)
        else:
            print(f"{count} * ", end="")
        N = N // count
    else:
        count += 1
```

7) Игра в угадайку - угадываем максимум за 10 попыток загаданное число от 1 до 1000 (если хотим от 1 до 1 000 000 надо расширить до 20 попыток)

```python
start = 0
end = 1000
count = 0

while count < 10:
    mid = (start + end) // 2
    print(mid)
    info = input()
    if info == "Больше":
        start = mid + 1
    elif info == "Меньше":
        end = mid - 1
    elif info == "Угадал!":
        count = 100
    count += 1
```

8) Очистка цифр на четных позициях числа

```python
N = int(input())
result = 0
count = 0
while N:
    if N % 10 % 2 != 0:
        result = result + (N % 10) * 10 ** count
        count += 1
    N = N // 10
print(result)
```
